```python
localities = list(codesignLocality['Locality'].str.lower())
```

```python
fidLocality['Area_match'] = fidLocality['Area'].apply(lambda x: x.upper() if x.lower() in localities else '')
fidLocality.head()
```

```python
len(fidLocality.loc[fidLocality['Area_match']==''])
```

```python
print(len(codesignLocality))
combinedLocalities = codesignLocality.merge(fidLocality, how='left', right_on=['Area_match','LGA'], left_on=['Locality','LGA']).drop('Area_match',axis=1)
print(len(combinedLocalities))
```

```python
combinedLocalities.head()
```

```python
fidUnmatchedLocalities = fidLocality.loc[fidLocality['Area_match']=='']
print(len(fidUnmatchedLocalities))
fidUnmatchedLocalities.head()
```

```python
combinedLocalities.to_csv('combinedLocalityProjections.csv', index=False)
fidUnmatchedLocalities.to_csv('unmatchedForecastIdLocalityProjections.csv', index=False)
```
