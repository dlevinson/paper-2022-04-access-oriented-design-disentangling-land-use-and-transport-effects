## Densities

```python
tzareas = pd.read_csv('./TPA_SPATIAL_TZ_NSW_2016_SHP/TZ_NSW_2016.csv')
tzareas.head()
```

```python
tzareas['Tz16_Area'].describe() # m2
```

```python
# Get TZ area (m.sq)

print(len(codesignTZ))
codesignTZ = codesignTZ.merge(tzareas[['TZ16_CODE','Tz16_Area']], how='left', left_on='tz16_code', right_on='TZ16_CODE').drop('TZ16_CODE', axis=1)
print(len(codesignTZ))
```

```python
codesignTZ.isnull().sum()
```

```python
# Density (per km.sq.)

codesignTZ['PopDensity_2016'] = codesignTZ['ERP_2016']*(10**6)/codesignTZ['Tz16_Area']
codesignTZ['PopDensity_2056'] = codesignTZ['ERP_2056']*(10**6)/codesignTZ['Tz16_Area']
codesignTZ['EmpDensity_2016'] = codesignTZ['EMP_2016']*(10**6)/codesignTZ['Tz16_Area']
codesignTZ['EmpDensity_2056'] = codesignTZ['EMP_2056']*(10**6)/codesignTZ['Tz16_Area']
codesignTZ['PopDensity_Diff'] = codesignTZ['PopDensity_2056']-codesignTZ['PopDensity_2016']
codesignTZ['EmpDensity_Diff'] = codesignTZ['EmpDensity_2056']-codesignTZ['EmpDensity_2016']
codesignTZ.head()
```

```python
codesignTZ.to_csv('TZ_Densities.csv', index=False)
```
