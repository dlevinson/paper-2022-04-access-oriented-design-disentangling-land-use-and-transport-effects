```python
import pandas as pd
import numpy as np
```

```python
codesignTZ = pd.read_csv('./CoDesign_Projections/TZprojections.csv', thousands = ',')
codesignTZ.head()
```

```python
print(len(codesignTZ))
codesignTZ = codesignTZ.loc[codesignTZ['tz16_code'].isnull()==False]
print(len(codesignTZ))
```

```python
codesignTZ.columns
```

```python
len(codesignTZ['SGS_CoDBC_SubPrecinct190924'].unique())
```

```python
codesignSubprecincts = codesignTZ[['SGS_CoDBC_SubPrecinct190924', 'ERP_2016', 'ERP_2056', 'EMP_2016',
                                   'EMP_2056']].groupby('SGS_CoDBC_SubPrecinct190924').sum().reset_index()
print(len(codesignSubprecincts))
codesignSubprecincts = codesignSubprecincts.merge(codesignTZ[['SGS_CoDBC_Precincts190924', 'SGS_CoDBC_SubPrecinct190924']].drop_duplicates(), how='left', on='SGS_CoDBC_SubPrecinct190924')
print(len(codesignSubprecincts))
codesignSubprecincts['ERP_2016'] = codesignSubprecincts['ERP_2016'].astype(int)
codesignSubprecincts['ERP_2056'] = codesignSubprecincts['ERP_2056'].astype(int)
codesignSubprecincts['EMP_2016'] = codesignSubprecincts['EMP_2016'].astype(int)
codesignSubprecincts['EMP_2056'] = codesignSubprecincts['EMP_2056'].astype(int)
codesignSubprecincts.head()
```

```python
print(codesignSubprecincts.loc[codesignSubprecincts['SGS_CoDBC_Precincts190924'].isin(['Aerotropolis', 'SWGA'])].drop('SGS_CoDBC_Precincts190924', axis=1).to_latex(index=False))
                         
                         
```

```python
codesignSubprecincts.to_csv('./CoDesign_Projections/CoDesign_Projections_SubPrecincts.csv', index=False)
```
