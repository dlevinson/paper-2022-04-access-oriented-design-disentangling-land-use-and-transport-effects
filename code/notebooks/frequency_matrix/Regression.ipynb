```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
import statsmodels.formula.api as smf
```

```python
accessdf = pd.read_csv('Access_LiverpoolLGA_Uniform.csv')
accessdf.head()
```

```python
accessdf['T_freqPerHr'] = accessdf['T_freq'].apply(lambda x: 0 if x==0 else 60/x).astype(int)
accessdf['L_freqPerHr'] = accessdf['L_freq'].apply(lambda x: 0 if x==0 else 60/x).astype(int)
accessdf.columns = ['locality','Jobs','cutoff','X','T_f','L_f','T_freq','L_freq']
accessdf.head()
```

```python
access15 = accessdf.loc[accessdf['cutoff']==15]
access30 = accessdf.loc[accessdf['cutoff']==30]
access45 = accessdf.loc[accessdf['cutoff']==45]
access60 = accessdf.loc[accessdf['cutoff']==60]
```

```python
reg15 = smf.ols('Jobs ~ T_freq + L_freq', data = access15).fit()
#reg15 = smf.ols('Jobs ~ T_freq + L_freq -1', data = access15).fit()
reg15.summary()
```

```python
reg30 = smf.ols('Jobs ~ T_freq + L_freq', data = access30).fit()
#reg30 = smf.ols('Jobs ~ T_freq + L_freq -1', data = access30).fit()
reg30.summary()
```

```python
reg45 = smf.ols('Jobs ~ T_freq + L_freq', data = access45).fit()
#reg45 = smf.ols('Jobs ~ T_freq + L_freq -1', data = access45).fit()
reg45.summary()
```

```python
reg60 = smf.ols('Jobs ~ T_freq + L_freq', data = access60).fit()
#reg60 = smf.ols('Jobs ~ T_freq + L_freq -1', data = access60).fit()
reg60.summary()
```
